```python
import requests

def calculate_distance(origin, destination, api_key):
    url = f"https://maps.googleapis.com/maps/api/distancematrix/json?origins={origin}&destinations={destination}&key={api_key}"
    response = requests.get(url)
    data = response.json()

    # Check if the API request was successful
    if data['status'] == 'OK':
        # Extract the distance value from the response
        distance = data['rows'][0]['elements'][0]['distance']['text']
        # Extract the duration value from the response
        duration = data['rows'][0]['elements'][0]['duration']['text']
        return distance, duration
    else:
        return None, None

# Replace API key
api_key = 'API'

service_point_addresses = [
    ('16928','Brünkendorf'),
    ('97483','Eltmann'),
    ('7426','Königsee'),
    ('27367','Hassendorf'),
    ('15926','Beesdau'),
    ('83371','Plattenberg'),
    ('72539','Pfronstetten'),
    ('77781','Biberach'),
    ('56290','Gödenroth'),
    ('23936','Gostorf')
]

client_addresses = [
    ('4157','Leipzig'),
    ('97483','Eltmann'),
    ('99706','Hachelbich'),
    ('1561','Kleinnaundorf'),
    ('4626','Heukewalde')
]

for service_point in service_point_addresses:
    origin = service_point[1]
    postal_code = service_point[0]
    origin_address = f"{postal_code}, {origin}"

    print(f"Calculating distances for Service Point: {origin_address}")
    print("=======================================")

    for client in client_addresses:
        destination = client[1]
        postal_code = client[0]
        destination_address = f"{postal_code}, {destination}"

        distance, duration = calculate_distance(origin_address, destination_address, api_key)

        if distance and duration:
            print(f"Distance from {origin_address} to {destination_address}: {distance}")
            print(f"Duration from {origin_address} to {destination_address}: {duration}")
        else:
            print(f"Unable to calculate distance for {destination_address}")

    print("\n")

```

```python
import requests
from openpyxl import Workbook

def calculate_distance(origin, destination, api_key):
    url = f"https://maps.googleapis.com/maps/api/distancematrix/json?origins={origin}&destinations={destination}&key={api_key}"
    response = requests.get(url)
    data = response.json()
# Replace  API key
api_key = 'API'

    if data['status'] == 'OK':
        # Extract the distance value from the response
        distance = data['rows'][0]['elements'][0]['distance']['text']
        # Extract the duration value from the response
        duration = data['rows'][0]['elements'][0]['duration']['text']
        return distance, duration
    else:
        return None, None



service_point_addresses = [
    ('16928','Brünkendorf'),
    ('97483','Eltmann'),
    ('7426','Königsee'),
    ('27367','Hassendorf'),
    ('15926','Beesdau'),
    ('83371','Plattenberg'),
    ('72539','Pfronstetten'),
    ('77781','Biberach'),
    ('56290','Gödenroth'),
    ('23936','Gostorf')
]

client_addresses = [
    ('4157','Leipzig'),
    ('97483','Eltmann'),
    ('99706','Hachelbich'),
    ('1561','Kleinnaundorf'),
    ('4626','Heukewalde')
]


wb = Workbook()
ws = wb.active

# Set the headers in the worksheet
ws['A1'] = 'Service Point'
ws['B1'] = 'Client'
ws['C1'] = 'Distance'
ws['D1'] = 'Duration'

row = 2  # Starting row for data

for service_point in service_point_addresses:
    origin = service_point[1]
    postal_code = service_point[0]
    origin_address = f"{postal_code}, {origin}"

    for client in client_addresses:
        destination = client[1]
        postal_code = client[0]
        destination_address = f"{postal_code}, {destination}"

        distance, duration = calculate_distance(origin_address, destination_address, api_key)

        if distance and duration:
            ws[f'A{row}'] = origin_address
            ws[f'B{row}'] = destination_address
            ws[f'C{row}'] = distance
            ws[f'D{row}'] = duration
        else:
            ws[f'A{row}'] = origin_address
            ws[f'B{row}'] = destination_address
            ws[f'C{row}'] = 'N/A'
            ws[f'D{row}'] = 'N/A'

        row += 1

# Save the workbook
wb.save('distance_results.xlsx')

```
